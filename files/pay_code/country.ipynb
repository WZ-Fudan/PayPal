```sql
CALL pp_monitor.drop_table('PP_OAP_SING_ZHEWEI_T.wz_case2_march_1');
create table PP_OAP_SING_ZHEWEI_T.wz_case2_march_1 as (
select trans_id
, rcvr_id
, pmt_start_date
, sndr_country
, case when brm_bad_tag_assigned = '~' then 1 else 0 end as is_good
from PP_OAP_SING_ZHEWEI_T.wz_case_march
)with data primary index(trans_id);
-------------------------------------------
CALL pp_monitor.drop_table('PP_OAP_SING_ZHEWEI_T.wz_case2_march_2');
create table PP_OAP_SING_ZHEWEI_T.wz_case2_march_2 as (
select rcvr_id
    , pmt_start_date
    , count(sndr_country) as tot_num
    , count(distinct sndr_country) as dis_num
    ,1 - sum(is_good) * 1.00000 / count(is_good) as bad_rate
from PP_OAP_SING_ZHEWEI_T.wz_case2_march_1
where trans_id > 0
group by 1, 2
)with data primary index(rcvr_id);
----------------------------------------------
call pp_monitor.drop_table('PP_OAP_SING_ZHEWEI_T.wz_case2_march_3');
create table PP_OAP_SING_ZHEWEI_T.wz_case2_march_3 as (
select rcvr_id
    , cast(avg(dis_num) as decimal(18,2)) as avg_ip
    , cast(avg(bad_rate) as decimal(18,5))as avg_bd
    from PP_OAP_SING_ZHEWEI_T.wz_case2_march_2
    group by 1
)with data primary index(rcvr_id);
```

```sql
--call pp_monitor.drop_table('PP_OAP_SING_ZHEWEI_T.wz_case2_merged_table');
create table PP_OAP_SING_ZHEWEI_T.wz_case2_merged_table as
(SELECT * FROM PP_OAP_SING_ZHEWEI_T.wz_country_data2
UNION
SELECT * FROM PP_OAP_SING_ZHEWEI_T.wz_case2_march_2)
with data primary index(rcvr_id);

--call pp_monitor.drop_table('PP_OAP_SING_ZHEWEI_T.wz_case2_window');
create table PP_OAP_SING_ZHEWEI_T.wz_case2_window as (
    select a.rcvr_id 
    , a.pmt_start_date
    , a.tot_num
    , a.dis_num
    , a.bad_rate
    , b.pmt_start_date as his_date
    , b.tot_num as his_tot_num
    , b.dis_num as his_dis_num
    , b.bad_rate as his_bad_rate
    from PP_OAP_SING_ZHEWEI_T.wz_case2_merged_table a
    left join PP_OAP_SING_ZHEWEI_T.wz_case2_merged_table b
    on a.rcvr_id = b.rcvr_id
    where a.pmt_start_date >= '2018-04-01'
    and b.pmt_start_date between a.pmt_start_date-30 and a.pmt_start_date-1
)with data primary index(rcvr_id);

--call pp_monitor.drop_table('PP_OAP_SING_ZHEWEI_T.wz_case2_window_result');
create table PP_OAP_SING_ZHEWEI_T.wz_case2_window_result as (
    select rcvr_id
    , pmt_start_date
    , tot_num
    , dis_num
    , bad_rate
    , cast(avg(his_tot_num) as decimal(18,5)) as avg_tot_num
    , cast(avg(his_dis_num) as decimal(18,5)) as avg_dis_num
    , cast(avg(his_bad_rate) as decimal(18,5)) as avg_bad_rate
    from PP_OAP_SING_ZHEWEI_T.wz_case2_window
    group by 1, 2, 3, 4, 5
)with data primary index(rcvr_id);
------------------------------------------
call pp_monitor.drop_table('PP_OAP_SING_ZHEWEI_T.wz_case2_contrast_filter');
create table PP_OAP_SING_ZHEWEI_T.wz_case_contrast2_filter AS(
select t1.rcvr_id
    , t1.pmt_start_date
    , t1.tot_num
    , t1.dis_num
    , t1.bad_rate
    , t1.tot_num * 1.00 / t1.dis_num as per_ip_num
    , t1.avg_tot_num
    , t1.avg_dis_num
    , t1.avg_bad_rate 
    , t1.avg_tot_num * 1.00 / t1.avg_dis_num as avg_per_ip_num
    from PP_OAP_SING_ZHEWEI_T.wz_case2_window_result t1
    left join PP_OAP_SING_ZHEWEI_T.template t2
    on t1.rcvr_id = t2.rcvr_id
    left join PP_OAP_SING_ZHEWEI_T.template1 t3
    on t1.rcvr_id = t3.rcvr_id
    where t3.rcvr_id is Null and 
        t2.rcvr_id is Null
)with data primary index(rcvr_id);
```
